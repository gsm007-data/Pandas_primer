```python
# =========================
# Formula References
# =========================
# Simple Interest:
#   Total Interest = P * R * T
#   EMI = (P + Total Interest) / N
#
# Compound Interest (EMI):
#   EMI = P * r * (1 + r)^n / ((1 + r)^n - 1)
#   where r = annual_rate / 12, n = total months

import matplotlib.pyplot as plt

def loan_calculator():
    principal = float(input("Enter loan amount: "))
    annual_rate = float(input("Enter annual interest rate (%): ")) / 100
    tenure_value = float(input("Enter loan tenure value: "))
    tenure_type = input("Is tenure in years or months? (y/m): ").lower()
    interest_type = input("Interest type? (simple/compound): ").lower()

    months = int(tenure_value * 12) if tenure_type == 'y' else int(tenure_value)
    monthly_rate = annual_rate / 12

    emi_list = []
    balance_list = []
    principal_list = []
    interest_list = []
    cumulative_interest_list = []

    if interest_type == "simple":
        total_interest = principal * annual_rate * (months / 12)
        total_payment = principal + total_interest
        emi = total_payment / months

        balance = principal
        cumulative_interest = 0

        for _ in range(months):
            interest = total_interest / months
            principal_paid = emi - interest
            balance -= principal_paid
            cumulative_interest += interest

            emi_list.append(emi)
            balance_list.append(max(balance, 0))
            principal_list.append(principal_paid)
            interest_list.append(interest)
            cumulative_interest_list.append(cumulative_interest)

    elif interest_type == "compound":
        emi = principal * monthly_rate * (1 + monthly_rate) ** months / ((1 + monthly_rate) ** months - 1)
        balance = principal
        cumulative_interest = 0

        for _ in range(months):
            interest = balance * monthly_rate
            principal_paid = emi - interest
            balance -= principal_paid
            cumulative_interest += interest

            emi_list.append(emi)
            balance_list.append(max(balance, 0))
            principal_list.append(principal_paid)
            interest_list.append(interest)
            cumulative_interest_list.append(cumulative_interest)

        total_interest = cumulative_interest
        total_payment = principal + total_interest

    else:
        return

    print(f"Monthly Payment (EMI): {emi_list[0]:.2f}")
    print(f"Total Payment: {total_payment:.2f}")
    print(f"Total Interest Paid: {total_interest:.2f}")

    months_range = range(1, months + 1)

    plt.figure()
    plt.plot(months_range, emi_list)
    plt.xlabel("Month")
    plt.ylabel("EMI")
    plt.title("EMI vs Month")
    plt.show()

    plt.figure()
    plt.plot(months_range, balance_list)
    plt.xlabel("Month")
    plt.ylabel("Remaining Balance")
    plt.title("Remaining Loan Balance vs Month")
    plt.show()

    plt.figure()
    plt.plot(months_range, principal_list, label="Principal")
    plt.plot(months_range, interest_list, label="Interest")
    plt.xlabel("Month")
    plt.ylabel("Amount")
    plt.title("Principal vs Interest per Month")
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(months_range, cumulative_interest_list)
    plt.xlabel("Month")
    plt.ylabel("Cumulative Interest")
    plt.title("Cumulative Interest Paid vs Time")
    plt.show()

    plt.figure()
    plt.pie([principal, total_interest], labels=["Principal", "Total Interest"], autopct="%1.1f%%")
    plt.title("Principal vs Total Interest")
    plt.show()

if __name__ == "__main__":
    loan_calculator()
```

```python
# =========================
# Formula References
# =========================
# Simple Interest:
#   Total Interest = P * R * T
#   EMI = (P + Total Interest) / N
#
# Compound Interest (EMI):
#   EMI = P * r * (1 + r)^n / ((1 + r)^n - 1)
#   where r = annual_rate / 12, n = total months

import csv
import matplotlib.pyplot as plt

def calculate_emi(principal, monthly_rate, months):
    return principal * monthly_rate * (1 + monthly_rate) ** months / ((1 + monthly_rate) ** months - 1)

def generate_amortization_table(principal, annual_rate, months):
    monthly_rate = annual_rate / 12
    emi = calculate_emi(principal, monthly_rate, months)

    table = []
    balance = principal

    for month in range(1, months + 1):
        interest = balance * monthly_rate
        principal_paid = emi - interest
        balance -= principal_paid

        table.append([
            month,
            round(emi, 2),
            round(principal_paid, 2),
            round(interest, 2),
            round(max(balance, 0), 2)
        ])

    return emi, table

def export_to_csv(table, filename="amortization_table.csv"):
    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Month", "EMI", "Principal Paid", "Interest Paid", "Remaining Balance"])
        writer.writerows(table)

def tenure_comparison(principal, annual_rate, tenures_years):
    emis = []
    total_interests = []

    for years in tenures_years:
        months = years * 12
        emi, table = generate_amortization_table(principal, annual_rate, months)
        total_payment = emi * months
        total_interest = total_payment - principal

        emis.append(emi)
        total_interests.append(total_interest)

    plt.figure()
    plt.plot(tenures_years, emis)
    plt.xlabel("Tenure (Years)")
    plt.ylabel("EMI")
    plt.title("Tenure vs EMI")
    plt.show()

    plt.figure()
    plt.plot(tenures_years, total_interests)
    plt.xlabel("Tenure (Years)")
    plt.ylabel("Total Interest")
    plt.title("Tenure vs Total Interest")
    plt.show()

def loan_calculator():
    principal = float(input("Enter loan amount: "))
    annual_rate = float(input("Enter annual interest rate (%): ")) / 100
    tenure_years = int(input("Enter loan tenure in years: "))

    months = tenure_years * 12

    emi, amort_table = generate_amortization_table(principal, annual_rate, months)
    export_to_csv(amort_table)

    for row in amort_table:
        print(row)

    tenures = [10, 15, 20]
    tenure_comparison(principal, annual_rate, tenures)

if __name__ == "__main__":
    loan_calculator()
```

```python
# Loan Calculator Program
# This program calculates the Monthly EMI, Total Payment, and Total Interest for a loan.

def loan_calculator():
    # Display header
    print("==== Loan Calculator ====")
    
    # --- Step 1: User Input ---
    # Input principal loan amount
    principal = float(input("Enter loan amount (‚Çπ): "))
    
    # Input annual interest rate in percentage
    annual_interest_rate = float(input("Enter annual interest rate (in %): "))
    
    # Ask user whether tenure is in Years or Months
    tenure_type = input("Is the loan tenure in (Y)ears or (M)onths? ").strip().lower()
    
    # Input tenure value
    tenure_value = float(input("Enter the loan tenure: "))
    
    # --- Step 2: Convert tenure to months ---
    if tenure_type == 'y':
        months = tenure_value * 12  # Convert years to months
    elif tenure_type == 'm':
        months = tenure_value  # Already in months
    else:
        print("Invalid tenure type. Please enter 'Y' or 'M'.")
        return
    
    # --- Step 3: Convert annual interest rate to monthly decimal rate ---
    monthly_interest_rate = annual_interest_rate / (12 * 100)
    
    # --- Step 4: EMI Calculation ---
    if monthly_interest_rate == 0:
        # If interest rate is 0%, EMI is simply principal divided by months
        emi = principal / months
    else:
        # Standard EMI formula for compound interest
        emi = (principal * monthly_interest_rate * (1 + monthly_interest_rate) ** months) / \
              ((1 + monthly_interest_rate) ** months - 1)
    
    # --- Step 5: Calculate total payment and total interest ---
    total_payment = emi * months  # Total amount to be paid
    total_interest = total_payment - principal  # Interest paid over principal
    
    # --- Step 6: Display Results ---
    print("\n===== Loan Summary =====")
    print(f"Loan Amount: ‚Çπ{principal:,.2f}")
    print(f"Annual Interest Rate: {annual_interest_rate:.2f}%")
    print(f"Loan Tenure: {months:.0f} months")
    print(f"Monthly EMI: ‚Çπ{emi:,.2f}")
    print(f"Total Payment: ‚Çπ{total_payment:,.2f}")
    print(f"Total Interest Paid: ‚Çπ{total_interest:,.2f}")
    print("=========================")


# Run the loan calculator
if __name__ == "__main__":
    loan_calculator()
```

```python
# Loan Calculator Program with Charts Export and Amortization Table
# Formulas:
# EMI = [P * r * (1 + r)^n] / [(1 + r)^n - 1]
# Total Payment = EMI * n
# Total Interest = Total Payment - P

import matplotlib.pyplot as plt
import pandas as pd

def calculate_loan_details(principal, annual_interest_rate, months):
    monthly_interest_rate = annual_interest_rate / (12 * 100)
    if monthly_interest_rate == 0:
        emi = principal / months
    else:
        emi = (principal * monthly_interest_rate * (1 + monthly_interest_rate) ** months) / \
              ((1 + monthly_interest_rate) ** months - 1)

    balance = principal
    emi_list, principal_component, interest_component, remaining_balance, cumulative_interest = [], [], [], [], []
    total_interest_paid = 0

    for m in range(1, months + 1):
        interest = balance * monthly_interest_rate
        principal_paid = emi - interest
        balance -= principal_paid
        total_interest_paid += interest

        emi_list.append(emi)
        principal_component.append(principal_paid)
        interest_component.append(interest)
        remaining_balance.append(balance if balance > 0 else 0)
        cumulative_interest.append(total_interest_paid)

    total_payment = emi * months
    total_interest = total_payment - principal

    amortization_data = {
        "Month": list(range(1, months + 1)),
        "EMI": emi_list,
        "Principal Paid": principal_component,
        "Interest Paid": interest_component,
        "Remaining Balance": remaining_balance
    }
    df = pd.DataFrame(amortization_data)
    
    return emi, total_payment, total_interest, df, remaining_balance, principal_component, interest_component, cumulative_interest


def loan_calculator():
    print("==== Loan Calculator ====")
    
    principal = float(input("Enter loan amount (‚Çπ): "))
    annual_interest_rate = float(input("Enter annual interest rate (in %): "))
    tenure_type = input("Is the loan tenure in (Y)ears or (M)onths? ").strip().lower()
    tenure_value = float(input("Enter the loan tenure: "))
    
    if tenure_type == 'y':
        months = int(tenure_value * 12)
    elif tenure_type == 'm':
        months = int(tenure_value)
    else:
        print("Invalid tenure type. Please enter 'Y' or 'M'.")
        return

    emi, total_payment, total_interest, df, remaining_balance, principal_component, interest_component, cumulative_interest = calculate_loan_details(
        principal, annual_interest_rate, months)

    print("\n===== Loan Summary =====")
    print(f"Loan Amount: ‚Çπ{principal:,.2f}")
    print(f"Annual Interest Rate: {annual_interest_rate:.2f}%")
    print(f"Loan Tenure: {months} months")
    print(f"Monthly EMI: ‚Çπ{emi:,.2f}")
    print(f"Total Payment: ‚Çπ{total_payment:,.2f}")
    print(f"Total Interest Paid: ‚Çπ{total_interest:,.2f}")
    print("=========================")
    
    print("\n===== Month-wise Amortization Table =====")
    print(df.head(12).to_string(index=False))
    df.to_csv("amortization_table.csv", index=False)
    print("\nAmortization table exported as 'amortization_table.csv'")

    print("\n===== Tenure Comparison =====")
    tenures_years = [5, 10, 15, 20, 25, 30]
    emi_values = []
    total_interests = []

    for t in tenures_years:
        m = t * 12
        _, _, ti, _, _, _, _, _ = calculate_loan_details(principal, annual_interest_rate, m)
        emi_temp = (principal * (annual_interest_rate / 1200) * (1 + (annual_interest_rate / 1200)) ** m) / \
                   ((1 + (annual_interest_rate / 1200)) ** m - 1)
        emi_values.append(emi_temp)
        total_interests.append(ti)

    comparison_df = pd.DataFrame({
        "Tenure (Years)": tenures_years,
        "Monthly EMI (‚Çπ)": emi_values,
        "Total Interest (‚Çπ)": total_interests
    })
    print(comparison_df.to_string(index=False))

    # === Create and Save All Charts to a Single File ===
    fig, axes = plt.subplots(2, 3, figsize=(14, 9))

    # EMI vs Month
    axes[0, 0].plot(df["Month"], df["EMI"], color='blue')
    axes[0, 0].set_title("EMI vs Month")
    axes[0, 0].set_xlabel("Month")
    axes[0, 0].set_ylabel("EMI Amount (‚Çπ)")
    axes[0, 0].grid(True)

    # Remaining Loan Balance vs Month
    axes[0, 1].plot(df["Month"], df["Remaining Balance"], color='red')
    axes[0, 1].set_title("Remaining Loan Balance vs Month")
    axes[0, 1].set_xlabel("Month")
    axes[0, 1].set_ylabel("Balance (‚Çπ)")
    axes[0, 1].grid(True)

    # Principal vs Interest per Month
    axes[0, 2].plot(df["Month"], df["Principal Paid"], color='green', label='Principal')
    axes[0, 2].plot(df["Month"], df["Interest Paid"], color='orange', label='Interest')
    axes[0, 2].set_title("Principal vs Interest per Month")
    axes[0, 2].set_xlabel("Month")
    axes[0, 2].set_ylabel("Amount (‚Çπ)")
    axes[0, 2].legend()
    axes[0, 2].grid(True)

    # Cumulative Interest Paid vs Time
    axes[1, 0].plot(df["Month"], cumulative_interest, color='purple')
    axes[1, 0].set_title("Cumulative Interest Paid vs Time")
    axes[1, 0].set_xlabel("Month")
    axes[1, 0].set_ylabel("Cumulative Interest (‚Çπ)")
    axes[1, 0].grid(True)

    # Pie Chart ‚Äì Principal vs Total Interest
    axes[1, 1].pie([principal, total_interest],
                   labels=["Principal", "Total Interest"],
                   autopct="%1.1f%%",
                   colors=['skyblue', 'lightcoral'],
                   startangle=90)
    axes[1, 1].set_title("Principal vs Total Interest")

    # Tenure Comparison Plot
    axes[1, 2].bar([str(t) for t in tenures_years], emi_values, color='teal')
    axes[1, 2].set_title("EMI vs Loan Tenure")
    axes[1, 2].set_xlabel("Tenure (Years)")
    axes[1, 2].set_ylabel("EMI (‚Çπ)")
    axes[1, 2].grid(True)

    plt.tight_layout()
    plt.savefig("loan_analysis_charts.png", dpi=300)
    plt.close()
    print("\nAll charts saved as 'loan_analysis_charts.png'")


if __name__ == "__main__":
    loan_calculator()
```

```python
# Simple To-Do List App
# Author: Goura Mohanty
# Description: A terminal-based To-Do List application that allows users to add, view, mark, and delete tasks.

def display_menu():
    print("\n==== To-Do List App ====")
    print("1. Add a new task")
    print("2. View all tasks")
    print("3. Mark a task as completed")
    print("4. Delete a task")
    print("5. Exit")


def view_tasks(tasks):
    if not tasks:
        print("\nNo tasks found!")
    else:
        print("\nYour Tasks:")
        for i, task in enumerate(tasks, start=1):
            status = "‚úì Completed" if task['completed'] else "‚úó Pending"
            print(f"{i}. {task['title']}  [{status}]")


def add_task(tasks):
    title = input("Enter new task: ").strip()
    if title:
        tasks.append({"title": title, "completed": False})
        print(f"Task '{title}' added successfully!")
    else:
        print("Task title cannot be empty.")


def mark_task_completed(tasks):
    view_tasks(tasks)
    if not tasks:
        return
    try:
        task_num = int(input("Enter task number to mark as completed: "))
        if 1 <= task_num <= len(tasks):
            tasks[task_num - 1]['completed'] = True
            print(f"Task '{tasks[task_num - 1]['title']}' marked as completed!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")


def delete_task(tasks):
    view_tasks(tasks)
    if not tasks:
        return
    try:
        task_num = int(input("Enter task number to delete: "))
        if 1 <= task_num <= len(tasks):
            removed_task = tasks.pop(task_num - 1)
            print(f"Task '{removed_task['title']}' deleted successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")


def todo_app():
    tasks = []
    while True:
        display_menu()
        choice = input("Enter your choice (1-5): ").strip()

        if choice == '1':
            add_task(tasks)
        elif choice == '2':
            view_tasks(tasks)
        elif choice == '3':
            mark_task_completed(tasks)
        elif choice == '4':
            delete_task(tasks)
        elif choice == '5':
            print("Exiting To-Do List App. Have a productive day!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")


if __name__ == "__main__":
    todo_app()
```

```python
# ==========================
# To-Do List App with Timestamps (Gradio Version)
# Author: Goura Mohanty
# Description: A beginner-friendly To-Do List app with full timestamp tracking.
# Features:
#   - Add Task (with Title + Description)
#   - View Tasks (shows timestamps)
#   - Mark as Completed (adds completion time)
#   - Delete Task (adds deletion time)
#   - Exit App
# ==========================

import gradio as gr
from datetime import datetime

# Global list to store tasks
# Each task is a dictionary containing title, description, timestamps, and status
tasks = []

# Helper function to get formatted current time
def current_time():
    """Returns current date and time as formatted string."""
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Function to add a new task
def add_task(title, description):
    """Add a new task with timestamps."""
    if not title.strip():
        return "‚ö†Ô∏è Please enter a valid task title."

    task = {
        "title": title.strip(),
        "description": description.strip() if description else "No description provided.",
        "completed": False,
        "created_at": current_time(),
        "completed_at": None,
        "deleted_at": None
    }
    tasks.append(task)
    return f"‚úÖ Task '{title}' added successfully at {task['created_at']}!"

# Function to view all tasks
def view_tasks():
    """Display all tasks with timestamps and status."""
    if not tasks:
        return "üìù No tasks found! Start by adding one."

    output = "üìã **Your Tasks:**\n\n"
    for i, task in enumerate(tasks, start=1):
        status = "‚úÖ Completed" if task['completed'] else "‚è≥ Pending"
        output += f"### {i}. {task['title']} ‚Äî *{status}*\n"
        output += f"üìù {task['description']}\n"
        output += f"üìÖ Created: {task['created_at']}\n"
        if task['completed_at']:
            output += f"üéØ Completed: {task['completed_at']}\n"
        if task['deleted_at']:
            output += f"üóëÔ∏è Deleted: {task['deleted_at']}\n"
        output += "\n"
    return output

# Function to mark a task as completed
def mark_completed(task_number):
    """Mark a task as completed and record completion time."""
    try:
        index = int(task_number) - 1
        if 0 <= index < len(tasks):
            if tasks[index]['completed']:
                return f"‚ÑπÔ∏è Task '{tasks[index]['title']}' is already marked as completed."
            tasks[index]['completed'] = True
            tasks[index]['completed_at'] = current_time()
            return f"üéØ Task '{tasks[index]['title']}' marked as completed at {tasks[index]['completed_at']}!"
        else:
            return "‚ö†Ô∏è Invalid task number."
    except ValueError:
        return "‚ö†Ô∏è Please enter a valid task number."

# Function to delete a task
def delete_task(task_number):
    """Delete a task and record deletion time."""
    try:
        index = int(task_number) - 1
        if 0 <= index < len(tasks):
            tasks[index]['deleted_at'] = current_time()
            removed = tasks.pop(index)
            return f"üóëÔ∏è Task '{removed['title']}' deleted at {removed['deleted_at']}!"
        else:
            return "‚ö†Ô∏è Invalid task number."
    except ValueError:
        return "‚ö†Ô∏è Please enter a valid task number."

# Main menu logic ‚Äî handles user choices
def menu(choice, title, description_or_number):
    """
    choice: Selected user action (Add, View, Mark, Delete, Exit)
    title: Task title (used when adding)
    description_or_number: Description (for Add) or task number (for Mark/Delete)
    """
    if choice == "Add Task":
        return add_task(title, description_or_number)
    elif choice == "View Tasks":
        return view_tasks()
    elif choice == "Mark as Completed":
        return mark_completed(description_or_number)
    elif choice == "Delete Task":
        return delete_task(description_or_number)
    elif choice == "Exit":
        return "üëã Exiting the To-Do App. Have a productive day!"
    else:
        return "‚ö†Ô∏è Invalid choice. Please select a valid option."


# -------------------------------
# Gradio User Interface
# -------------------------------
with gr.Blocks() as todo_app:
    gr.Markdown("## üßæ To-Do List App with Timestamps\nTrack your tasks with created, completed, and deleted times.")

    # Dropdown menu for user action
    menu_choice = gr.Dropdown(
        ["Add Task", "View Tasks", "Mark as Completed", "Delete Task", "Exit"],
        label="Select an action"
    )

    # Task title input
    task_title = gr.Textbox(label="Task Title", placeholder="Example: Complete Python project")

    # Task description or task number input
    task_description_or_number = gr.Textbox(
        label="Task Description or Task Number",
        placeholder="Example: Work on the loan calculator module OR 2"
    )

    # Output display box
    output_box = gr.Markdown(label="Output")

    # Submit button
    submit_btn = gr.Button("Submit")

    # Link the button click event to the menu logic
    submit_btn.click(
        fn=menu,
        inputs=[menu_choice, task_title, task_description_or_number],
        outputs=output_box
    )

# Run the Gradio app
if __name__ == "__main__":
    todo_app.launch()
```
