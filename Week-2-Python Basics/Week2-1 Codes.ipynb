# 2. Lists

## 2.1 Creating Lists

```python
# Empty list
my_list = []

# List with different data types
my_list = [1, 'apple', 3.5]

# List created using the list() constructor with range()
numbers = list(range(1, 6))   # Result: [1, 2, 3, 4, 5]

# Creating a list from another sequence (string → list of characters)
letters = list("Python")      # Result: ['P', 'y', 't', 'h', 'o', 'n']
```

## 2.2 Accessing List Elements

```python
# A list of fruits
fruits = ['apple', 'banana', 'cherry']

# Accessing the first element (index 0)
print(fruits[0])    # Output: apple

# Accessing the last element using negative indexing
print(fruits[-1])   # Output: cherry
```

## 2.3 Modifying List Elements

```python
# Initial list
numbers = [1, 2, 3, 4]

# Change an element (replace value at index 1)
numbers[1] = 9          # Result: [1, 9, 3, 4]

# Add a new item at the end
numbers.append(5)       # Result: [1, 9, 3, 4, 5]

# Insert an item at a specific position (index 0)
numbers.insert(0, 0)    # Result: [0, 1, 9, 3, 4, 5]

# Remove the first occurrence of a value
numbers.remove(3)       # Result: [0, 1, 9, 4, 5]

# Remove item at a specific index (index 2)
del numbers[2]          # Result: [0, 1, 4, 5]

# pop an item from the end
last_item = numbers.pop()  # last_item = 5, numbers = [0, 1, 4]

# clearing the entire list
numbers.clear()          # Result: []
```

## 2.4 Sorting and Reversing

```python
colors = ['red', 'blue', 'green']

# Sort the list alphabetically - 
colors.sort()          # Result: ['blue', 'green', 'red']

sorted_colors = sorted(colors)  # New sorted list
print(sorted_colors)           # Output: ['blue', 'green', 'red']

# Reverse the list order
colors.reverse()       # Result: ['red', 'green', 'blue']
```

## 2.5 Counting and Finding

```python
numbers = [1, 2, 3, 2, 1]

# Count how many times 2 appears
count = numbers.count(2)     # Result: 2

# Find the index of the first occurrence of 3
index = numbers.index(3)     # Result: 2
```

## 2.6 Adding and Extending

```python
list1 = [1, 2, 3]
list2 = [4, 5]

# Extend list1 by adding all elements of list2
list1.extend(list2)    # Result: [1, 2, 3, 4, 5]
```

## 2.7 Slicing Lists

```python
# Original list
letters = ['a', 'b', 'c', 'd', 'e']

# Get elements from index 1 to 3 (stop index is excluded)
sliced_list = letters[1:4]      # Result: ['b', 'c', 'd']

# Get all elements from index 2 onwards
subset = letters[2:]            # Result: ['c', 'd', 'e']

# Get every second element using step
every_second = letters[::2]     # Result: ['a', 'c', 'e']

# Reverse a list using slicing
reversed_list = letters[::-1]   # Result: ['e', 'd', 'c', 'b', 'a']
```

## 2.8 List mutability

# 3. Tuple

## 3.1 How to Create a Tuple

```python
# Empty tuple
empty_tuple = ()

# Tuple with different data types
mixed_tuple = (1, 'apple', 3.5)

# Single-item tuple (must include a trailing comma)
single_item_tuple = (5,)     # Without the comma, (5) is just an integer

# Tuple created without parentheses (tuple packing)
coordinates = 10, 20, 30     # Result: (10, 20, 30)

# Using the tuple() constructor
vowels = tuple('aeiou')      # Result: ('a', 'e', 'i', 'o', 'u')
```

## 3.3 Tuple Operations

# 4. Set

## 4.1 How to Create a Set

```python
# Empty set
empty_set = set()          # NOT {} — {} creates an empty dictionary

# Set with elements
fruits = {'apple', 'banana', 'cherry'}

# Set with different data types
mixed_set = {1, 'apple', 3.5}

# Creating a set from another iterable (duplicates removed)
numbers = set([1, 2, 2, 3, 4, 4])   # Result: {1, 2, 3, 4}
```

## 4.2 Properties of Sets

```python
# ----------------------
# Unordered (no indexing)
# ----------------------
fruits = {'apple', 'banana', 'cherry'}

# This will raise an error because sets don't support indexing:
# print(fruits[0])   # ❌ TypeError


# ----------------------
# Unique Elements
# ----------------------
numbers = set([1, 2, 2, 3, 3, 3])
print(numbers)       # Output: {1, 2, 3}


# ----------------------
# Mutable but with Restrictions
# ----------------------

# Works because tuples are immutable
s = {1, 2, (3, 4)}

# Fails because lists are mutable
# s = {1, 2, [3, 4]}   # ❌ TypeError: unhashable type: 'list'
```

## 4.3 Common Operations on Sets

```python
# Starting set
fruits = {'apple', 'banana', 'cherry'}

# Adding elements
fruits.add('orange')        # Result: {'apple', 'banana', 'cherry', 'orange'}

# Removing elements (error if not found)
# fruits.remove('banana')   # Works
# fruits.remove('pear')     # ❌ KeyError

# Remove without error
fruits.discard('pear')      # No error if not found

# Checking membership
print('apple' in fruits)    # Output: True

# Length of a set
print(len(fruits))          # Example output: 3


# Other methods:
# pop()      → removes and returns an arbitrary element
# clear()    → removes all elements
# copy()     → returns a shallow copy
```

## 4.4 Mathematical Operations on Sets

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

# ----------------------
# Union
# ----------------------
print(set1 | set2)               # Output: {1, 2, 3, 4, 5}
print(set1.union(set2))          # Same result


# ----------------------
# Intersection
# ----------------------
print(set1 & set2)               # Output: {3}
print(set1.intersection(set2))   # Same result


# ----------------------
# Difference
# ----------------------
print(set1 - set2)               # Output: {1, 2}
print(set1.difference(set2))     # Same result


# ----------------------
# Symmetric Difference
# ----------------------
print(set1 ^ set2)                      # Output: {1, 2, 4, 5}
print(set1.symmetric_difference(set2))  # Same result

```

## 4.5 Use Cases of Sets

```python
# ============================================================
#  Use Cases of Sets
# ============================================================

# ----------------------
# Removing Duplicates
# ----------------------
items = ['apple', 'banana', 'apple', 'orange']
unique_items = set(items)        # Output: {'orange', 'banana', 'apple'}


# ----------------------
# Membership Testing
# ----------------------
fruits = {'apple', 'banana', 'cherry'}
print('banana' in fruits)        # Output: True


# ----------------------
# Finding Unique or Common Elements
# ----------------------
visitors_day1 = {'Alice', 'Bob', 'Charlie'}
visitors_day2 = {'Bob', 'David', 'Eve'}

# Common visitors (intersection)
returning = visitors_day1 & visitors_day2   # Output: {'Bob'}

# All visitors (union)
all_visitors = visitors_day1 | visitors_day2
```

# 5. Dictionary

## 5.1 How to Create a Dictionary

```python
# Empty dictionary
empty_dict = {}            # OR dict()

# Dictionary with items
person = {'name': 'Alice', 'age': 25, 'city': 'New York'}

# Using dict() constructor
car = dict(make='Toyota', model='Corolla', year=2020)

# Dictionary from list of tuples
items = [('apples', 5), ('bananas', 3)]
fruit_count = dict(items)     # Output: {'apples': 5, 'bananas': 3}

# Dictionary comprehension
squares = {x: x**2 for x in range(6)}
# Output: {0:0, 1:1, 2:4, 3:9, 4:16, 5:25}
```

## 5.2 Accessing and Modifying Dictionary Elements

```python
# ----------------------
# Accessing Values
# ----------------------
person = {'name': 'Alice', 'age': 25}

print(person['name'])          # Output: Alice
print(person.get('city'))      # Output: None
print(person.get('city', 'Unknown'))   # Output: Unknown


# ----------------------
# Adding & Updating
# ----------------------
# Add new key-value pair
person['city'] = 'New York'

# Update an existing value
person['age'] = 26

# Update multiple values at once
person.update({'age': 27, 'job': 'Engineer'})


# ----------------------
# Removing Items
# ----------------------
# Remove a specific key-value pair
del person['city']

# Remove and return a value
job = person.pop('job')         # Returns 'Engineer'

# Remove and return last inserted item (Python 3.7+)
last_item = person.popitem()

# Remove all items
person.clear()
```

## 5.3 Dictionary Methods

```python
person = {'name': 'Alice', 'age': 25}

# Getting keys as a view object
keys = person.keys()           # dict_keys(['name', 'age'])

# Getting values as a view object
values = person.values()       # dict_values(['Alice', 25])

# Getting key-value pairs as a view object
items = person.items()         # dict_items([('name', 'Alice'), ('age', 25)])

# Copy a dictionary
person_copy = person.copy()

# Create a new dictionary with default values
template = dict.fromkeys(['name', 'age', 'city'], 'Unknown')
# {'name': 'Unknown', 'age': 'Unknown', 'city': 'Unknown'}
```

```python
# View objects auto-update when dictionary changes
person = {'name': 'Alice'}
k = person.keys()

print(k)              # dict_keys(['name'])

person['age'] = 25
print(k)              # dict_keys(['name', 'age']) → auto-updated
```

## 5.4 Nested Dictionaries

```python
# Employee records with nested information
employees = {
    'emp1': {
        'name': 'Alice',
        'age': 25,
        'skills': ['Python', 'SQL'],
        'contact': {
            'email': 'alice@example.com',
            'phone': '555-1234'
        }
    },
    'emp2': {
        'name': 'Bob',
        'age': 30,
        'skills': ['Java', 'C++'],
        'contact': {
            'email': 'bob@example.com',
            'phone': '555-5678'
        }
    }
}

# Accessing nested data
print(employees['emp1']['name'])                 # Output: Alice
print(employees['emp2']['contact']['email'])     # Output: bob@example.com

# Safely accessing deep keys using get()
email = employees.get('emp3', {}).get('contact', {}).get('email')
print(email)                                      # Output: None (emp3 doesn't exist)
```

## 5.5 Limitations of Dictionaries

```python
# ----------------------
# Key Restrictions
# ----------------------
# ✔ Works: tuple as a dictionary key (immutable)
d = {(1, 2): 'tuple key'}
print(d[(1, 2)])        # Output: 'tuple key'

# ❌ Error: lists are mutable → cannot be keys
# d = {[1, 2]: 'list key'}          # TypeError

# ❌ Error: sets are mutable → cannot be keys
# d = {{1, 2}: 'set key'}           # TypeError


# ----------------------
# Memory Usage
# ----------------------
# (No code needed — conceptual)


# ----------------------
# Order Considerations
# ----------------------
# Python 3.7+ preserves insertion order:
demo = {}
demo['a'] = 1
demo['b'] = 2
demo['c'] = 3
print(demo)     # {'a': 1, 'b': 2, 'c': 3}
```
